### 2. Посчитайте совокупный объем средств у клиентов в банке - сумма по всем балансам. 

```python
sql("""
    SELECT 
        Cust_ID,
        Cheq_Bal + Sav_Bal + Debt_Bal AS 'Sum Balance'
    FROM bank;
""")
```

### 3. Выведите только тех клиентов, которые приобрели ипотеку. 

```python
sql("""
   SELECT Cust_ID
   FROM bank
   WHERE Purchase == 1;
""")
```

### 4. Какой совокупный объем средств в среднем у клиентов, которые приобрели ипотеку и у клиентов, которые её не приобретали? 

```python
sql("""
    SELECT AVG(Cheq_Bal + Sav_Bal + Debt_Bal) AS 'Sum Balance'
    FROM bank
    GROUP BY Purchase;
""")
```

### 5. Есть ли разница в среднем по каждому из типов баланса у клиентов, которые приобрели ипотеку и у клиентов, которые её не приобретали? 

```python
sql("""
    SELECT 
        AVG(Cheq_Bal) AS mean_cheq_bal,
        AVG(Sav_Bal) AS mean_sav_bal,
        AVG(Debt_Bal) AS mean_debt_bal
    FROM bank
    GROUP BY Purchase;
""")
```

### 6. Выберите самого возрастного получателя ипотеки. 

```python
sql("""
    SELECT *
    FROM bank
    WHERE Purchase == 1
    ORDER BY Age DESC
    LIMIT 1;
""")
```

### 7. Выберите клиента, который раньше всех получил ипотеку.

### 8. Создайте любую базу данных в sqlite3, состоящую из 5 колонок, одна из колонок обязательно должна быть типа REAL. (Названия колонок любые, какие захотите) 

```python
con = connect('patients_data.db')

con.execute("""
    CREATE TABLE patients (
    id INTEGER,
    first_name STRING,
    last_name STRING,
    age INTEGER,
    weight REAL,
    lab_id INTEGER
    );
""")

con.commit()

con.close()
```

```python
# проверяем содержимое созданной базы данных
con = connect('patients_data.db')
cur = con.cursor()
cur.execute("SELECT * FROM patients")
data = cur.fetchall()
con.close()
data
```

### 9. Добавьте 3 полностью заполненные строки в вашу БД. 

```python
con = connect('patients_data.db')
cur = con.cursor()

data = [
    ('1', 'Ivan', 'Ivanov', '56', '78.9', '101'),
    ('2', 'Elena', 'Fedorova', '34', '58.1', '106'),
    ('3', 'Kira', 'Petrova', '60', '64.6', '110')
]

cur.executemany("""
    INSERT INTO patients
    VALUES (?, ?, ?, ?, ?, ?);
""", data)

con.commit()
con.close()
```

```python
# проверяем содержимое созданной базы данных
con = connect('patients_data.db')
cur = con.cursor()
cur.execute("SELECT * FROM patients")
data = cur.fetchall()
con.close()
data
```

### 10. Удалите строку из колонки типа REAL (из пункта 8), в которой значение признака среднее (то есть не минимальное и не максимальное). Напрямую указывать значение из этой строки нельзя. 

```python
con = connect('patients_data.db')
cur = con.cursor()

cur.execute("""
    DELETE FROM patients
    WHERE weight <> (SELECT MIN(weight) FROM patients)
        AND weight <> (SELECT MAX(weight) FROM patients);
""")

con.commit()
con.close()
```
